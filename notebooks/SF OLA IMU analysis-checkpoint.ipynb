```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import Artemis IMU Data

```python
still_data = pd.read_csv('artemis data - still.csv')
x_mov = pd.read_csv('artemis data - x axis.csv')
y_mov = pd.read_csv('artemis data - y-axis.csv')
```

```python
still_data = still_data.drop(['rtcDate','rtcTime'], axis = 1)
x_mov = x_mov.drop(['rtcDate','rtcTime'], axis = 1)
y_mov = y_mov.drop(['rtcDate','rtcTime'], axis = 1)
```

```python
still_data
```

```python
x_mov
```

```python
y_mov
```

# Calculate average accuracy error

```python
Avg_error_still = np.mean(abs(still_data - np.mean(still_data))/abs(np.mean(still_data)) * 100)
Avg_error_x = np.mean(abs(x_mov - np.mean(x_mov))/abs(np.mean(x_mov)) * 100)
Avg_error_y = np.mean(abs(y_mov - np.mean(y_mov))/abs(np.mean(y_mov)) * 100)
```

```python
Avg_error_still, Avg_error_x, Avg_error_y
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('Average Accuracy Error of Still Artemis')
x = Avg_error_still.index
y = Avg_error_still.values
ax.bar(x,y,color = 'pink')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('Average Accuracy Error of X-moving Artemis')
x = Avg_error_x.index
y = Avg_error_x.values
ax.bar(x,y,color = 'cyan')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.set_title('Average Accuracy Error of y-moving Artemis')
x = Avg_error_y.index
y = Avg_error_y.values
ax.bar(x,y,color = 'orange')
plt.show()
```
